```python
import matplotlib.pyplot as plt
%matplotlib notebook
import numpy as np

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

fig_width_pt = 240.0  # Get this from LaTeX using \showthe\columnwidth
inches_per_pt = 1.0/72.27               # Convert pt to inches
golden_mean = (np.sqrt(5)-1.0)/2.0         # Aesthetic ratio
fig_width = fig_width_pt*inches_per_pt  # width in inches
fig_height =fig_width*golden_mean       # height in inches
fig_size = [fig_width,fig_height]
```

```python
data = np.load('output/extinctions.npy')
```

```python
fig= plt.figure(figsize=(8,9))
for i in range(len(data)):
    plt.subplot(5,4,i+1)
    ax = plt.gca()
   

    d = data[i].copy()
    if ~(d == 0).all():
        plt.hist(data[i])
        #plt.xlabel('$a_v$')
        #plt.ylabel('Occurence')
        ax.text(.7,.7,'C' + str(i).zfill(2),transform=ax.transAxes)
    else:
        frame = plt.gca()
        frame.axes.get_xaxis().set_ticks([])
        frame.axes.get_yaxis().set_ticks([])
        ax.text(.4,.45,'C' + str(i).zfill(2),transform=ax.transAxes)
    
fig.text(0.5, 0.04, '$a_v$', ha='center',fontsize =20)
fig.text(0.04, 0.5, 'Occurence', va='center', rotation='vertical',fontsize =20)    
plt.subplots_adjust(wspace=.3)
plt.savefig('all_dust_hist.pdf')
```
