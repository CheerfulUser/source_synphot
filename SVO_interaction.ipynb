```python
import xml.etree.ElementTree as et
import requests
import os
from svo_filters import svo

def downloadFilter(experiment, band):
    savepath = os.path.join( os.path.split(svo.__file__)[0], 'data', 'filters')
    print("Downloading filter " + band)
    url = 'http://svo2.cab.inta-csic.es/theory/fps/fps.php?ID=' + experiment + '/' + band
    page = requests.get(url)
    #check to make sure this is a valid filter
    root = et.fromstring(page.content)
    if(root.findall('INFO')[0].attrib['value'] == "OK"):
        with open(os.path.join(savepath, band), 'wb') as file:
            file.write(page.content)
    else:
        raise FileNotFoundError("Filter " + band + " for experiment " + experiment + " was not found in the SVO database.")
```
