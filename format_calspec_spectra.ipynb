```python
%matplotlib notebook
import matplotlib.pyplot as plt
from astropy.io import fits
import numpy as np
import pandas as pd
from glob import glob
```

```python
files = glob('../../data/current_calspec/*.fits')
```

```python
files
```

```python
for file in files:
    a = fits.open(file)[1]
    wave = a.data['WAVELENGTH']
    flux = a.data['FLUX']
    
    df = pd.DataFrame(columns=['wave','flux'])
    df.wave = wave
    df.flux = flux
    name = file.split('.fits')[0]+'.dat'
    print(name) 
    df.to_csv(name,index=False)
```
