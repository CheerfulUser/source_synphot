```python
import astropy.table as at
import os
import source_synphot.io as io
import pandas as pd
import numpy as np
```

```python
source_table_file = os.path.join('sources', 'test.txt')
```

```python
source_table_file = io.get_pkgfile(source_table_file)
```

```python
source_table = at.Table.read(source_table_file, format='ascii')
```

```python
source_table = source_table.to_pandas()
```

```python
source_table
```

```python
ps1 = pd.read_csv('../WDdata/wds_ps1.csv')
```

```python
source_table.specname.values
```

```python
source_table.keys()
```

```python
list(ps1.keys())
```

```python
for i in range(len(ps1)):
    for j in range(len(source_table)):
        print (ps1.objid.iloc[i],source_table.specname.iloc[j])
        if ps1.objid.iloc[i].lower() in source_table.specname.iloc[j].lower():
            source_table['g'].iloc[j] = np.round(ps1['gMeanPSFMag'].iloc[i],2)
            source_table['r'].iloc[j] = np.round(ps1['rMeanPSFMag'].iloc[i],2)
            source_table['i'].iloc[j] = np.round(ps1['iMeanPSFMag'].iloc[i],2)
            source_table['z'].iloc[j] = np.round(ps1['zMeanPSFMag'].iloc[i],2)
```

```python
source_table.to_csv('test.txt', header=True, index=False, sep='\t')
```
